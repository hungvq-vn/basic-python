```python
''' viết 1 chương trình chuyển đổi nhiệt độ từ độ C sang độ F và ngược lại '''

def c_to_f(celsius):
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit

def f_to_c(fahrenheit):
    celsius = (fahrenheit - 32) * 5/9
    return celsius

#check 
enter = input ( "Nhập giá trị của đơn vị đo nhiệt độ (C/F) : ")
temp = float(input("Nhập giá trị nhiệt độ: ")) 
if enter.upper() == 'C' :
    print(f" {temp} độ C tương đương với {c_to_f(temp):.2f} độ F")
elif enter.upper() == 'F':
    print(f" {temp} độ F tương đương với {f_to_c(temp):.2f} độ C")
else:
    print("Đơn vị đo không hợp lệ. Vui lòng nhập C hoặc F.")
```

```python
# CLASS METHOD  , STATIC METHOD : nó như là 1 hàm bình thường , không cần truyền bất cứ tham số hay đối số nào vào , kiểu nó có cho hay hoặc nó có chỉ để mục đích thông báo thôi 

''' @classmethod dùng để thao tác với class một cách linh hoạt (dynamic) thông qua cls
→ nên khi class con gọi, NÓ SẼ TÁC ĐỘNG ĐÚNG CLASS ĐÓ, không hard-code class gốc

Nó giúp:
-> không làm ảnh hưởng sai class khác khi có kế thừa

-> tái sử dụng code cho mọi class con

-> kiểm soát logic (validate, xử lý trước khi thay đổi biến class)'''

# class method dùng để khi các class khác kế thừa thì nó tác động vào đúng class đang gọi nó (như việc cls để tạo các object riêng ) - đồng thời nó giúp lọc các điều kiện đầu vào khi thay đổi cái variable class  . CƠ BẢN : classmethod chỉ xử lý với các variable class thôi , còn biến của object thì cứ self

''' VÍ DỤ DỄ HIỂU '''

class A:
    x = 10

    @classmethod
    def set_x(cls, value):  # cls riêng để khi class A được kế thừa thì nó không làm thay đổi x của A 
        cls.x = value

class B(A):
    pass 

B.set_x(20)  # dù có thay đổi ở đây thì nó cũng chỉ thay đổi ở B chứ không ảnh hưởng gì A


print(f"Đây là x của A : {A.x}")
print(f"Đây là x của B : {B.x}")
```

```python
''' xử lý trường hợp nhiều đối số bằng cách dùng *args  '''

class employee:
  def __init__(self, name, salary):
        self.name = name
        self.salary = salary
  @staticmethod  # cái này tính bên ngoài cho hay nên có thể dùng staticmethod được , nó không cần đối số hay gì 
  def salary_sum(*full_list_employee) :
      sum = 0
      for nv in full_list_employee :
          sum += nv.salary 
      return sum 

# Chạy thử
nv1 = employee("Hùng", 5000)
nv2 = employee("Lan", 3000)
nv3 =  employee("Quốc" , 6000)

full_cost = employee.salary_sum(nv1,nv2,nv3)
print(f" Full salary cost of employee is : {full_cost}")
```

```python
''' Khi nào cần dùng tới @property  : 
- Khi cần bảo vệ dữ liệu (Validation): Như ví dụ của bạn, không muốn ai đó vô tình gán lương bị âm, tuổi bị âm, hoặc email thiếu chữ @.

- Khi muốn tính toán động (Computed Property): Ví dụ bạn có self.chieu_dai và self.chieu_rong. Bạn có thể dùng @property cho hàm dien_tich (chỉ trả về chieu_dai * chieu_rong) mà không cần tạo thêm biến self.dien_tich.

-- Khi cấu trúc lại code (Refactoring): Giả sử ban đầu bạn dùng biến public nv1.salary. Sau này dự án lớn lên, bạn cần thêm điều kiện khi đổi lương. Thay vì phải đi sửa hàng ngàn dòng code nv1.salary = ... trong toàn dự án thành nv1.set_salary(...), bạn chỉ cần thêm @property và @setter vào trong class( với tên của hàm được chuyển thành thuộc tính PHẢI  với cái tên biến cần được sửa ngoài kia ). Các dòng code cũ ở bên ngoài vẫn hoạt động bình thường! '''

''' Cái @property và cái @....setter là như này : 
- nó sẽ biến 1 hàm thành 1 thuộc tính và có thể gọi như 1 thuộc tính . và khi ta gọi @proper ty với tên hàm là ta tạo 1 hàm với 1 thuộc tính không thể bị thay đổi , chỉ có thể xem chứ không thể đổi .(thực ra vẫn có thể đổi nếu thiết lập đổi ở trong cái property đó luôn ) . sau đó mình gọi cái @setter đó thì thực ra là mình lấy cái hàm trên property mình ném thêm tính năng cho nó thôi . khi được property thì bái hàm đó không khác gì 1 biến bình thường NHƯNG bên trong nó phức tạp hơn '''

class employee:
    def __init__(self, name, salary):
        self.name = name
        self.__salary = salary  # khoá lại bằng 2 gạch phía dưới , bên ngoài kh hể truy cập , chỉ có thể truy cập ở trong hàm thôi , nếu muốn được truy cập như 1 biến thì phải dùng property thêm đồ chơi
                
    @property
    def salary(self):
        return self.__salary

    @salary.setter
    def salary(self, so_tien_moi):
        if so_tien_moi < 0:
            print(f"Lỗi: Số tiền {so_tien_moi}$  không hợp lệ")
        else:
            self.__salary = so_tien_moi
            print(f"Cập nhật lương thành công số tiền {so_tien_moi}!")

nv1 = employee("Hùng", 5000)

print(nv1.salary) 

nv1.salary = -2000 

nv1.salary = 8000 

```

```python
''' Khi muốn tính toán động (Computed Property): Ví dụ bạn có self.chieu_dai và self.chieu_rong. Bạn có thể dùng @property cho hàm dien_tich (chỉ trả về chieu_dai * chieu_rong) mà không cần tạo thêm biến self.dien_tich. '''
# VÍ DỤ CỤ THỂ 

class HinhChuNhat:
    def __init__(self, chieu_dai, chieu_rong):
        self.chieu_dai = chieu_dai
        self.chieu_rong = chieu_rong
        self.dien_tich = chieu_dai * chieu_rong
print("---------- VỚI TRƯỜNG HỢP KHÔNG DÙNG @PROPERTY ----------")
hcn = HinhChuNhat(5, 3)
print(f"Diện tích hình chữ nhật khi CHƯA thay đổi các đối số chiều dài, rộng là : {hcn.dien_tich} ")
hcn.chieu_dai = 10
print(f"Diện tích hình chữ nhật khi ĐÃ thay đổi các đối số chiều dài, rộng là : {hcn.dien_tich}\n ")

#------------------------------------------------------------------------------------------------

''' Nguyên nhân gây ra lỗi logic
Tính toán 1 lần duy nhất: Khi bạn tạo đối tượng bằng lệnh hcn = HinhChuNhat(5, 3), hàm __init__ được gọi. Lúc này, Python lấy 5 nhân 3 và gán giá trị 15 vào biến self.dien_tich. Từ thời điểm này, self.dien_tich chỉ là một biến lưu trữ con số 15 tĩnh.

Không tự động cập nhật: Khi bạn gọi hcn.chieu_dai = 10, thuộc tính chieu_dai đã thay đổi. Tuy nhiên, thuộc tính dien_tich không hề có cơ chế nào để biết rằng chiều dài đã thay đổi để tự tính lại. Nó vẫn giữ nguyên con số 15 được cấp lúc ban đầu.

Đó là lý do dù bạn đổi chiều dài thành 10 hay 100, lệnh print(hcn.dien_tich) bên dưới vẫn mãi mãi in ra 15.'''

#-------------------------------------------------------------------------------------------------------
print("---------- VỚI TRƯỜNG HỢP DÙNG @PROPERTY ----------")
class HinhChuNhatt:
    def __init__(self, dai, rong):
        self.dai = dai
        self.rong = rong

    @property
    def dien_tichh(self):
        return self.dai * self.rong
hcn = HinhChuNhatt(5, 3)
print(f"Diện tích hình chữ nhật khi CHƯA thay đổi các đối số chiều dài, rộng là : {hcn.dien_tichh} ")

hcn.dai = 10
print(f"Diện tích hình chữ nhật khi ĐÃ thay đổi các đối số chiều dài, rộng là : {hcn.dien_tichh} ")
```
